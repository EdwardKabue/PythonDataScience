```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("england-premier-league-matches-2018-to-2019-stats.csv", parse_dates=True)
```

```python
df.head()
```

```python
df["red_cards"] = df.home_team_red_cards + df.away_team_red_cards
```

```python
#Set style to darkgrid.
sns.set_style("darkgrid")
```

```python
#Plotting 3D graph.
ax = plt.axes(projection="3d")
ax.scatter(df.index,df["red_cards"],df["total_goal_count"])
ax.set_xlabel("Index")
ax.set_ylabel("red_cards")
ax.set_zlabel("total_goal_count")
```
