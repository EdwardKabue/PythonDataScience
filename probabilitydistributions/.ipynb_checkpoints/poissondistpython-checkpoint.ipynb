```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
#random variable representing the number of restaurants
X = [0, 1, 2, 3, 4, 5]
#mean number of restaurants located within 10 miles is 2
lmbda = 2
```

```python
#probability values
poisson_pd = poisson.pmf(X, lmbda)
```

```python
#plot the probability distribution
fig, ax = plt.subplots(1, 1 , figsize=(8,6))
ax.plot(X, poisson_pd, 'bo', ms=8, label="poisson_pmf")
plt.ylabel("Probability", fontsize="18")
plt.xlabel("X = No. of restuarants Vs Probability", fontsize="18")
ax.vlines(X, 0, poisson_pd, colors="b", lw=5, alpha=0.5)
```
